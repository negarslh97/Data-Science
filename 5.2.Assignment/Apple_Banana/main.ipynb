```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apples = pd.DataFrame({
    'length': np.random.normal(6, 0.6 , 600),
    'width': np.random.normal(6, 0.6 , 600)
    })

bananas = pd.DataFrame({
    'length': np.random.normal(8, 0.6 , 600),
    'width': np.random.normal(4, 0.6 , 600)
    })

apples['fruit_type'] = 'apple'
bananas['fruit_type'] = 'banana'
```

```python
fruits = pd.concat([bananas, apples])
print(fruits.head())
```

```python
plt.scatter(x=apples['length'], y=apples['width'], label='apples' , c='red')
plt.scatter(x=bananas['length'], y=bananas['width'], label='bananas' , c='yellow')
plt.legend()
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Distribution of Apples and Banans')
plt.show()
```

```python
apples_mean_length = apples['length'].mean()
apples_mean_width = apples['width'].mean()

bananas_mean_length = bananas['length'].mean()
bananas_mean_width = bananas['width'].mean()
```

```python
labels = ['Length', 'Width']
apples_means = [apples_mean_length, apples_mean_width]
bananas_means = [bananas_mean_length, bananas_mean_width]
```

```python
x = np.arange(len(labels))
width = 0.35
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, apples_means, width, label='Apples', color='red')
rects2 = ax.bar(x + width/2, bananas_means, width, label='Bananas', color='yellow')

ax.set_ylabel('Values')
ax.set_title('Average Length and Width of Fruits')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```python
data = [apples['length'], bananas['length'], apples['width'], bananas['width']]
labels = ['Apples Length', 'Bananas Length', 'Apples Width', 'Bananas Width']
colors=['red' , 'yellow']

plt.boxplot(data, labels=labels)
plt.title('Box Plot of Length and Width for Apples and Bananas')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.hist(apples['width'], bins=30, alpha=0.7, label='Apples Width', color='red')
plt.hist(bananas['width'], bins=30, alpha=0.7, label='Bananas Width', color='yellow')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.title('Width Distribution')
plt.legend()

plt.subplot(1, 2, 1)
plt.hist(apples['length'], bins=30, alpha=0.7, label='Apples Length', color='red')
plt.hist(bananas['length'], bins=30, alpha=0.7, label='Bananas Length', color='yellow')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.title('Length Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
fig, (ax, cax) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]}, figsize=(10, 6))

h = ax.hist2d(fruits['length'], fruits['width'], bins=50, cmap='jet')
ax.set_xlabel('Length')
ax.set_ylabel('Width')
ax.set_title('2D Histogram of Length and Width')
plt.colorbar(h[3], cax=cax)

plt.show()
```

```python
fea = 'L/W'
fruits[fea] = fruits['length'] / fruits['width']

apples_data = fruits.loc[fruits['fruit_type'] == 'apple', fea]
bananas_data = fruits.loc[fruits['fruit_type'] == 'banana', fea]

plt.hist(apples_data, bins=30, alpha=0.7, label='Apples Width', color='red')
plt.hist(bananas_data, bins=30, alpha=0.7, label='Bananas Width', color='yellow')

plt.legend()
plt.title('L/W Distribution')
plt.xlabel('L/W ratio')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
areas = 'area'
fruits[areas] = fruits['length']*fruits['width']

apples_data = fruits.loc[fruits['fruit_type'] == 'apple', areas]
bananas_data = fruits.loc[fruits['fruit_type'] == 'banana', areas]

plt.hist(apples_data, bins=30, alpha=0.7, histtype='step', lw=4 , label='Apples Width', color='red')
plt.hist(bananas_data, bins=30, alpha=0.7, histtype='step', lw=4 , label='Bananas Width', color='yellow')

plt.legend()
plt.title('Area Distribution')
plt.xlabel('Area')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
circumference = 'circumference'
fruits[circumference] = 2*(fruits['length']+fruits['width'])

apples_data = fruits.loc[fruits['fruit_type'] == 'apple', circumference]
bananas_data = fruits.loc[fruits['fruit_type'] == 'banana', circumference]

plt.hist(apples_data, bins=30, alpha=0.7, histtype='step', lw=4 , label='Apples Width', color='red')
plt.hist(bananas_data, bins=30, alpha=0.7, histtype='step', lw=4 , label='Bananas Width', color='yellow')

plt.legend()
plt.title('circumference Distribution ')
plt.xlabel('Circumference')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
