```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('california_cities.csv')
```

```python
print(df.head())
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
plt.scatter(df['latd'], df['longd'] ,
            s=df['area_total_km2'],
            alpha=0.5 ,
            linewidths=0 , 
            c=np.log10(df['population_total']),
            cmap='gist_rainbow')
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.colorbar(label='log$_{10}$(population)')
plt.xlabel('Longitude')
plt.ylabel('Longitude')
plt.title('Area and Population of California Cities')
plt.show()
```
