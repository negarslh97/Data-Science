Calculate the Percentage of Egual Goals

```python
def is_equalizer(row):
    if row['Home team'] == 'FC Barcelona' or row['Home team'] == 'Argentina':
        return row['Score Home Team - When Messi Scored'] == row['Score Away Team - When Messi Scored']
    else:
        return row['Score Away Team - When Messi Scored'] == row['Score Home Team - When Messi Scored']
```

```python
df['Is Equalizer'] = df.apply(is_equalizer, axis=1)
df
```

```python
equalizer_goals = df['Is Equalizer'].sum()
total_goals = len(df)
equalizer_percentage = (equalizer_goals / total_goals) * 100
equalizer_percentage
```
