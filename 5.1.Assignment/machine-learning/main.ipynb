```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = [120,80,98,170,160,90,88,120,140,110]
```

```python
meanData = np.mean(data)

print(meanData)
```

```python
medianData = np.median(data)

print(meanData)
```

```python
modeData = stats.mode(data)

print(modeData)
```

```python
standardDeviation = np.std(data)

print(standardDeviation)
```

```python
varianceData = np.var(data)

print(varianceData)
```

```python
percentileData = np.percentile(data , 75)

print(percentileData)
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel , color="orange")
plt.show()
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline) , c='red')
plt.show()
```

```python
from sklearn.metrics import r2_score

print(r2_score(y, mymodel(x)))
```
