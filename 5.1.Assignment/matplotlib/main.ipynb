```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "purple")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1 , color = "purple")
plt.show()
```

```python
data = np.random.normal(300, 100, 250)

plt.hist(data , color = "brown")
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.legend(title = "Fruits" , loc="upper right", bbox_to_anchor=(1, 1))
plt.show() 
```
